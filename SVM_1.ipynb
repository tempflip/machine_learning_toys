```python
import numpy as np
from sklearn import datasets
iris = datasets.load_iris()
```

```python
iris.keys()
```

```python
print(iris['feature_names'])
print(iris['target_names'])
```

```python
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
X = iris['data']
Y = iris['target']
X[:,:2]
# filtering
x = []
y = []
for i in range(len(X)):
    if Y[i] == 0 : continue
    x.append(X[i][0:2])
    y.append(Y[i])
X = np.array(x)
Y = np.array(y)

Y
```

```python
from sklearn import svm
clf = svm.SVC(gamma=10, C=100)
clf.fit(X,Y)
clf.get_params()
clf.score(X,Y)
```

```python
contour_gran = 4
Z = np.zeros((10 * contour_gran, 10 * contour_gran))

for i in range(Z.shape[0]):
    for j in range(Z.shape[1]):
        Z[i, j] = clf.predict([j / contour_gran,i / contour_gran])
        pass

print(Z)
```

```python
plt.figure(figsize=(10,10))
plt.contourf(np.linspace(0,10, 10 * contour_gran), np.linspace(0,10, 10 * contour_gran) ,Z, 1, cmap=plt.get_cmap('cool'))
#plt.contourf(Z)
plt.xlim([3,10])
plt.ylim([0,6])
plt.scatter(X[:,0], X[:,1], c = Y)
```
